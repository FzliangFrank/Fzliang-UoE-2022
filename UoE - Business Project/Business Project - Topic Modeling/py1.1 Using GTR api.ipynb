## 1. First Api callings

`response.json()` is a `dict` object which is already a json object one you can use as if it is a list. 

`json.load()` and `json.dumps()` are two function that let you transfer from str to json and json to string. 

The "project" is a list of things. Within list is another json project. So basically you have to select those elements.

The maximum data you can get use this method is 20 (19 as python counts from zero)

for a full list of the example: https://gtr.ukri.org/gtr/api/examples

```python
parameters = {
    "term":"Artificial Intelligence",
    "size":100
}

url = "https://gtr.ukri.org/gtr/api/projects.json"

response = requests.get(url, params = parameters)
print(response.status_code)
```

```python
i = 1
x = response.json()
print(x['project'][i]['title'],"\n")
print(x['project'][i]['abstractText'])
```

## 2. refine related to AI

Filter are applyed after `json`. `sf=fu.sd&so=A`: filter by starting date of funding. Cannot use this to extract fundings. 

Anything includes term "Artificial Intelligence", so it could be anything, does not has to be a research project..

```python
parameters = {
    "q":"artificial intelligence",
    "s":100,
    "p":1,
    "f":"pro.rt"
}

url = "https://gtr.ukri.org/gtr/api/projects.json?sf=fu.sd&so=A&pro.am=true"

response = requests.get(url, params = parameters)
print(response.status_code)
```

```python
i = 1
x = response.json()
status = x['project'][i]['status']

try:
    time = x['project'][i]['links']['link'][-1]['start']
    date = datetime.datetime.fromtimestamp(
        int(time/1000)
    ).strftime('%Y-%m-%d')
    
    print("title: ", x['project'][i]['title'], "\n")

    print("lead funder: ", x['project'][i]['leadFunder'])
    print("start date: ", date)
    print("status: ", status)
    print("\n","abstract: ", x['project'][i]['abstractText'])
    
except: 
    print("title: ", x['project'][i]['title'], "\n")

    print("lead funder: ", x['project'][i]['leadFunder'])
    print("no starting date")
    print("status: ", status)
    print("\n","abstract: ", x['project'][i]['abstractText'])
```

## 3. Search only in project topic that contains artificial intelligence ⭐️

   Note above that if you just search term that contains Artifician inteligence it does not covers only research topic, some are very general like PhD ship and so. If you analysis those together you might commite pseudorepliation that is some fundings goes to university then goes to researcher, you might find it confusing how funding received are higher than that provided (becasue you have counted those fundings twice). 
   
   Add filters after a question mark after `gtr~json`.`sf` is for sorting method. `so` soring by acending `A` or decending `D`. Here search only in project topic `f`(search field)`pro.rt`(reference research topic). Access full list of filter in [configuration part]().

```python
parameters = {
    "q":"artificial intelligence",
    "s":20,
    "p":1,
    "f":"pro.rt" # search in project topic 
}
url = "https://gtr.ukri.org/gtr/api/projects.json?sf=pro.sd&so=A" # sort by project start date Decending (D) Acending(A)

response = requests.get(url, params = parameters)
print(response.status_code)
x = response.json()
# explore first layer
pd.DataFrame.from_dict(x['project']).info()
```
