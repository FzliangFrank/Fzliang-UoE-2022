```python
import pandas as pd
import seaborn as sns 
import numpy as np
import matplotlib.pyplot as plt
```

```python
from uk_covid19 import Cov19API
england_only = [
    'areaType=nation',
    'areaName=England'
]
cases_and_deaths = {
    "date": "date",
    "cumTest" :"cumTestsByPublishDate",
    "cumCase" : "cumCasesBySpecimenDate",
    "dailyCase" : "newCasesBySpecimenDate",
    "dailyTest" :"newTestsByPublishDate"
}
api = Cov19API(filters=england_only, structure=cases_and_deaths)
df = api.get_dataframe()
#print(df)
#df = pd.DataFrame(data)
#df = pd.read_csv(iterator = True, data)
df["date"] = pd.to_datetime(df["date"])
df.index = df["date"]
```

```python
df["infection rate (daily)"] = df["dailyCase"]/df["dailyTest"]
df["infection rate (cum)"] = df['cumCase']/df["cumTest"]
```

```python
sns.relplot(x = "date", y = "infection rate (daily)", data = df
            ,height = 8, aspect=13/8.27)
```

```python
sns.relplot(x = "date", y = "infection rate (cum)", data = df
            ,height = 8, aspect=13/8.27)
```

```python
sns.relplot(x = "date", y = "dailyCase", data = df,height = 8, aspect=13/8.27)
sns.relplot(x = "date", y = "dailyTest", data = df,height = 8, aspect=13/8.27)
```
